```python
import pandas as pd
```

#nav_history Cleaning

```python
dt_2=pd.read_csv('02_nav_history.csv')
```

```python
print(dt_2.describe())
print(dt_2.info())
```

```python
#parse dates into datetime
dt_2.date=pd.to_datetime(dt_2.date)
```

```python
dt_2.info()
```

```python
#sort by amfi_code+date
dt_2=dt_2.sort_values(by=['amfi_code','date'])
```

```python
#forward fill for missing nav
dt_2.nav=dt_2.groupby('amfi_code').nav.ffill()
```

```python
#drop duplicates
dt_2=dt_2.drop_duplicates()
```

```python
#Validate nav>0
dt_2=dt_2[dt_2.nav>0]
```

```python
dt_2.to_csv('02_nav_history_cleaned.csv',index=False)
```

#scheme_performance cleaning

```python
dt_7=pd.read_csv('07_scheme_performance.csv')
```

```python
dt_7.info()
```

```python
#Validate returns to numeric
returns=['return_1yr_pct','return_3yr_pct','return_5yr_pct']

for col in returns:
  dt_7[col]=pd.to_numeric(dt_7[col],errors='coerce')
```

```python
dt_7.info()
```

```python
#Flag anomalies
dt_7['anomaly']=dt_7[returns].isnull().any(axis=1)
```

```python
#check expense ratio range
dt_7=dt_7[(dt_7.expense_ratio_pct>=0.1) & (dt_7.expense_ratio_pct<=2.5)]
```

```python
dt_7.to_csv('07_scheme_performance_cleaned.csv',index=False)
```

#investor_transactions cleaning

```python
dt_8=pd.read_csv('08_investor_transactions.csv')
```

```python
dt_8.head()
```

```python
dt_8.info()
```

```python
#Standardize transaction type
dt_8.transaction_type=dt_8.transaction_type.str.upper().str.strip()
```

```python
dt_8=dt_8.rename(columns={'transaction_date':'date'})
```

```python
#Validate amount
dt_8=dt_8[dt_8.amount_inr>0]
```

```python
#Fix Date formats
dt_8.transaction=pd.to_datetime(dt_8.date,errors='coerce')
```

```python
dt_8.info()
```

```python
#Validate KYC
valid_kyc=['Verified','Pending','Rejected']
dt_8=dt_8[dt_8.kyc_status.isin(valid_kyc)]
```

```python
dt_8.to_csv('08_investor_transactions_cleaned.csv',index=False)
```
